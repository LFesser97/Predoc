```python
# import packages

import os
import cv2

from IPython.display import Image
```

```python
# set path directories & load image

path = os.getcwd()

image_path = os.path.join(path, "images")

image_name = 'TK1930_371_b269.jpeg'

img = cv2.imread(os.path.join(image_path, image_name))

print(img.shape)

rect_img = cv2.rectangle(img, (490, 320), (1970, 2350), (0, 255, 0), 10)

cv2.imwrite(os.path.join(image_path, 'rect_' + image_name), rect_img)

Image(filename = os.path.join(image_path, 'rect_' + image_name)) 
```

```python
# downsize image

fac = 0.2 # resize factor

res_img = cv2.resize(img, (0,0), fx = fac, fy = fac)

# cv2.imwrite(os.path.join(image_path, 'res_' + image_name), res_img)

print(res_img.shape)

rect_res_img = cv2.rectangle(res_img, (98, 64), (394, 588), (0, 255, 0), 2)

cv2.imwrite(os.path.join(image_path, 'rect_res_' + image_name), rect_res_img)

Image(filename = os.path.join(image_path, 'rect_res_' + image_name)) 
```
